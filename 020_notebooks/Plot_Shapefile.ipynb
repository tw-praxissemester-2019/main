```python
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = gpd.read_file("../010_data/wc2/shp/1/file.shp")
```

```python
print(len(df))
```

```python
# Hier schauen, ob die Legende nur Stufen zeigen kann
# Und Achsenbeschriftungen
fig, ax = plt.subplots(figsize=(20, 10))
df.plot(column='DN', ax=ax, legend=True)
```

```python
p = Point(-119.913,38.5592)
```

```python
dn=df[df.contains(p)]['DN'].iloc[0]
```

```python
print(dn)
```

```python
for index, row in df.iterrows():
    if(p.within(row['geometry'])):
        print("found");
        break;
```
